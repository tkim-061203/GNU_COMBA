```python
import json, re, tempfile, subprocess, os
import numpy as np

def module_extraction(example_code):
    with tempfile.NamedTemporaryFile(delete=False, suffix=".v") as fp:
        fp.write(example_code.encode(encoding='utf-8'))
        fp.close()
        sub_run = subprocess.run(['verible-verilog-format', fp.name], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        
        # example_code = sub_run.stdout.decode("utf-8")
        with open(fp.name, 'r') as file:
            example_code = file.read()
    code_lines = example_code.splitlines()
    regex_stage = 0
    regexes = [r'^.*module\s+', r'^.*module$']
    module_def_span = np.array([None, None, None, None])
    num_indexs = 0
    comment_stage = 0
    multiline_comment = False
    for i in range(len(code_lines)):
        line = code_lines[i]
        num_indexs += (i > 0) # plus previous \n
        if comment_stage == 0:
            line_for_comment = str(line)
            comment_ranges = []
            comment_num_indexs = 0
            while True:
                double_slash_index = line_for_comment.find("//")
                double_slash_index = [double_slash_index, double_slash_index]
                
                single_slash_index = line_for_comment.find("/*")
                single_slash_index = [single_slash_index, single_slash_index]
                
                single_slash_end_index = line_for_comment.rfind("*/")
                single_slash_end_index = [single_slash_end_index, single_slash_end_index]

                #
                # Handle for case of multi-line comment
                if multiline_comment:
                    if (double_slash_index[0] != -1 and single_slash_end_index[0] != -1):
                        if (double_slash_index[0] < single_slash_end_index[0]) or (double_slash_index[0] - single_slash_end_index[0] == 1):
                            double_slash_index[0] = -1
                        # else: No else!!!                            
                else:
                    if double_slash_index[0] != -1 and single_slash_index[0] != -1: # fix all cases of comment!!
                        if double_slash_index[0] < single_slash_index[0]:
                            single_slash_index[0] = -1
                        else:
                            double_slash_index[0] = -1
                    if (double_slash_index[0] != -1 and single_slash_end_index[0] != -1):
                        if double_slash_index[0] < single_slash_end_index[0]:
                            single_slash_end_index[0] = -1
                    if single_slash_end_index[0] > -1 and single_slash_index[0] > -1 and (single_slash_end_index[0] - single_slash_index[0] == 1):
                        single_slash_end_index[0] = single_slash_end_index[1] = -1
                #
                # handle for start comment
                # start_comment_index = double_slash_index if double_slash_index != None else single_slash_index if single_slash_index != None else None
                if multiline_comment:
                    if single_slash_end_index[0] != -1:
                        multiline_comment = False

                        len_of_old_line = len(line_for_comment)
                        comment_ranges.append([0, single_slash_end_index[0] + 1])
                        comment_ranges[-1][0] += comment_num_indexs
                        comment_ranges[-1][1] += comment_num_indexs
                        
                        line_for_comment = line_for_comment[single_slash_end_index[0]+2:]
                        comment_num_indexs += len_of_old_line - len(line_for_comment)
                    else:
                        len_of_old_line = len(line_for_comment)
                        if len_of_old_line:
                            comment_ranges.append([0, len_of_old_line - 1])
                            comment_ranges[-1][0] += comment_num_indexs
                            comment_ranges[-1][1] += comment_num_indexs
                            
                            line_for_comment = line_for_comment[:0]
                            comment_num_indexs += len_of_old_line - len(line_for_comment)
                        break
                
                if not multiline_comment:
                    if double_slash_index[0] != -1:
                        len_of_old_line = len(line_for_comment)
                        comment_ranges.append([double_slash_index[0], len_of_old_line - 1])
                        comment_ranges[-1][0] += comment_num_indexs
                        comment_ranges[-1][1] += comment_num_indexs
                        
                        line_for_comment = line_for_comment[:double_slash_index[0]]
                        comment_num_indexs += len_of_old_line - len(line_for_comment)
                    if single_slash_index[0] != -1 and single_slash_end_index[0] == -1:
                        multiline_comment = True

                        len_of_old_line = len(line_for_comment)
                        comment_ranges.append([single_slash_index[0], len_of_old_line - 1])
                        comment_ranges[-1][0] += comment_num_indexs
                        comment_ranges[-1][1] += comment_num_indexs
                        
                        line_for_comment = line_for_comment[:single_slash_index[0]]
                        comment_num_indexs += len_of_old_line - len(line_for_comment)
                        
                    elif single_slash_index[0] != -1 and single_slash_end_index[0] != -1:
                        len_of_old_line = len(line_for_comment)
                        comment_ranges.append([single_slash_index[0], single_slash_end_index[0]+1])
                        comment_ranges[-1][0] += comment_num_indexs
                        comment_ranges[-1][1] += comment_num_indexs
                        
                        line_for_comment = line_for_comment[:single_slash_index[0]] + line_for_comment[single_slash_end_index[0]+2:]
                        comment_num_indexs += len_of_old_line - len(line_for_comment)
                        
                if double_slash_index[1] != -1 or single_slash_index[1] != -1 or single_slash_end_index[1] != -1:
                    continue
                else:
                    break
                
        # if len(comment_ranges):
        #     print(comment_ranges, i)

        section_start_idx = 0
        cur_section = ''

        if len(comment_ranges): # sort for correct order
            comment_ranges.sort(key=lambda item: item[0])
            
        while section_start_idx < len(line):
            old_section_start_idx = section_start_idx
            if len(comment_ranges):
                cur_section = line[section_start_idx:comment_ranges[0][0]]
                section_start_idx += comment_ranges[0][1] + 1
                comment_ranges.pop(0)
            else:
                cur_section = line[section_start_idx:]
                section_start_idx += len(cur_section)
        
            #
            if regex_stage == 0:
                for regex in regexes:
                    module_match = re.match(regex, cur_section)
                    if module_match != None:
                        regex_stage += 1
                        span = module_match.span()
                        module_def_span = np.vstack((module_def_span, [span[0] + num_indexs, span[1] + num_indexs, None, None]))
                        break
            if regex_stage == 1:
                semicon_pos = cur_section.find(";")
                if semicon_pos != -1:
                    module_def_span[-1][-2] = num_indexs + semicon_pos + 1
                    regex_stage += 1
            if regex_stage == 2:
                endmodule_pos = cur_section.find("endmodule")
                if endmodule_pos != -1:
                    module_def_span[-1][-1] = num_indexs + endmodule_pos + 1
                    regex_stage = 0
    
            num_indexs += section_start_idx - old_section_start_idx
            
    if len(module_def_span.shape) < 2:
        error_txt = f"Module span is all null {fp.name}, {examples}, {len(example_code)}: {example_code}"
        
        raise Exception(error_txt)
        
    module_def_span = np.delete(module_def_span, 0, 0)
    module_wrapper = ""
    module_definition = []
    module_output_code = []
    for span in module_def_span:
        if None in span:
            raise Exception(f"Span Error!! {examples},  {fp.name}, {len(example_code)}: {example_code}")
        module_definition += [example_code[span[0]:span[2]]]
        module_output_code += [example_code[span[2]:span[3]+8]]
        
    module_definition_with_upper_comments = []
    for i in range(len(module_def_span)):
        span = module_def_span[i]
        if i == 0:
            span[0] = 0
        else:
            span[0] = module_def_span[i-1][-1] + 1
        module_definition_with_upper_comments += [example_code[span[0]:span[2]]]

    subprocess.run(['rm', '-f', fp.name])

    return (module_definition, module_output_code, module_definition_with_upper_comments)
code = '//-----------------------------------------------------------------------------\n// The confidential and proprietary information contained in this file may\n// only be used by a person authorised under and to the extent permitted\n// by a subsisting licensing agreement from ARM Limited.\n//\n//            (C) COPYRIGHT 2008-2009 ARM Limited.\n//                ALL RIGHTS RESERVED\n//\n// This entire notice must be reproduced on all copies of this file\n// and copies of this file may only be made by a person if such person is\n// permitted to do so under the terms of a subsisting license agreement\n// from ARM Limited.\n//\n//      SVN Information\n//\n//      Checked In          : $Date: 2008-12-04 12:22:21 +0000 (Thu, 04 Dec 2008) $\n//\n//      Revision            : $Revision: 95812 $\n//\n//      Release Information : Cortex-M0-AT510-r0p0-01rel0\n//-----------------------------------------------------------------------------\n\n// ----------------------------------------------------------------------------\n// Cortex-M0 Integration Implementation level Vector Capture Testbench \n// ----------------------------------------------------------------------------\n//\n// This block generates vectors by snooping on pins at the Cortex-M0 Integration\n// Implementation level. The vectors are written to the file\n// CORTEXM0INTEGRATIONIMP_vectors.crf\n// ----------------------------------------------------------------------------\n\nmodule cm0ik_cortexm0integrationimp_capture (FCLK, SCLK, HCLK, DCLK, PORESETn, DBGRESETn, HRESETn, SWCLKTCK, nTRST, HRDATA, HREADY, HRESP, SWDITMS, TDI, DBGRESTART, EDBGRQ, NMI, IRQ, RXEV, STCALIB, STCLKEN, IRQLATENCY, ECOREVNUM, SLEEPHOLDREQn, WICENREQ, CDBGPWRUPACK, SE, RSTBYPASS, SYSRETAINn, SYSISOLATEn, SYSPWRDOWN, DBGISOLATEn, DBGPWRDOWN, HADDR, HBURST, HMASTLOCK, HPROT, HSIZE, HTRANS, HWDATA, HWRITE, HMASTER, CODENSEQ, CODEHINTDE, SPECHTRANS, SWDO, SWDOEN, TDO, nTDOEN, DBGRESTARTED, HALTED, TXEV, LOCKUP, SYSRESETREQ, GATEHCLK, SLEEPING, SLEEPDEEP, WAKEUP, WICSENSE, SLEEPHOLDACKn, WICENACK, CDBGPWRUPREQ, SYSPWRDOWNACK, DBGPWRDOWNACK);\ninput FCLK;\ninput SCLK;\ninput HCLK;\ninput DCLK;\ninput PORESETn;\ninput DBGRESETn;\ninput HRESETn;\ninput SWCLKTCK;\ninput nTRST;\ninput [31:0] HRDATA;\ninput HREADY;\ninput HRESP;\ninput SWDITMS;\ninput TDI;\ninput DBGRESTART;\ninput EDBGRQ;\ninput NMI;\ninput [31:0] IRQ;\ninput RXEV;\ninput [25:0] STCALIB;\ninput STCLKEN;\ninput [7:0] IRQLATENCY;\ninput [27:0] ECOREVNUM;\ninput SLEEPHOLDREQn;\ninput WICENREQ;\ninput CDBGPWRUPACK;\ninput SE;\ninput RSTBYPASS;\ninput SYSRETAINn;\ninput SYSISOLATEn;\ninput SYSPWRDOWN;\ninput DBGISOLATEn;\ninput DBGPWRDOWN;\ninput [31:0] HADDR;\ninput [2:0] HBURST;\ninput HMASTLOCK;\ninput [3:0] HPROT;\ninput [2:0] HSIZE;\ninput [1:0] HTRANS;\ninput [31:0] HWDATA;\ninput HWRITE;\ninput HMASTER;\ninput CODENSEQ;\ninput [2:0] CODEHINTDE;\ninput SPECHTRANS;\ninput SWDO;\ninput SWDOEN;\ninput TDO;\ninput nTDOEN;\ninput DBGRESTARTED;\ninput HALTED;\ninput TXEV;\ninput LOCKUP;\ninput SYSRESETREQ;\ninput GATEHCLK;\ninput SLEEPING;\ninput SLEEPDEEP;\ninput WAKEUP;\ninput [33:0] WICSENSE;\ninput SLEEPHOLDACKn;\ninput WICENACK;\ninput CDBGPWRUPREQ;\ninput SYSPWRDOWNACK;\ninput DBGPWRDOWNACK;\n`ifndef ARM_CM0IK_SKIPVECNUM\n`define ARM_CM0IK_SKIPVECNUM 0\n`endif // ARM_CM0IK_SKIPVECNUM\n`ifdef ARM_CM0IK_XVAL\n`define ARM_CM0IK_XVALUE "L"\n`else\n`define ARM_CM0IK_XVALUE "H"\n`endif // ARM_CM0IK_XVAL\nreg  vFCLK;\nreg  vSCLK;\nreg  vHCLK;\nreg  vDCLK;\nreg  vPORESETn;\nreg  vDBGRESETn;\nreg  vHRESETn;\nreg  vSWCLKTCK;\nreg  vnTRST;\nreg [31:0] vHRDATA__31_0;\nreg  vHREADY;\nreg  vHRESP;\nreg  vSWDITMS;\nreg  vTDI;\nreg  vDBGRESTART;\nreg  vEDBGRQ;\nreg  vNMI;\nreg [31:0] vIRQ__31_0;\nreg  vRXEV;\nreg [25:0] vSTCALIB__25_0;\nreg  vSTCLKEN;\nreg [7:0] vIRQLATENCY__7_0;\nreg [27:0] vECOREVNUM__27_0;\nreg  vSLEEPHOLDREQn;\nreg  vWICENREQ;\nreg  vCDBGPWRUPACK;\nreg  vSE;\nreg  vRSTBYPASS;\nreg  vSYSRETAINn;\nreg  vSYSISOLATEn;\nreg  vSYSPWRDOWN;\nreg  vDBGISOLATEn;\nreg  vDBGPWRDOWN;\nreg [31:0] vHADDR__31_0;\nreg [2:0] vHBURST__2_0;\nreg  vHMASTLOCK;\nreg [3:0] vHPROT__3_0;\nreg [2:0] vHSIZE__2_0;\nreg [1:0] vHTRANS__1_0;\nreg [31:0] vHWDATA__31_0;\nreg  vHWRITE;\nreg  vHMASTER;\nreg  vCODENSEQ;\nreg [2:0] vCODEHINTDE__2_0;\nreg  vSPECHTRANS;\nreg  vSWDO;\nreg  vSWDOEN;\nreg  vTDO;\nreg  vnTDOEN;\nreg  vDBGRESTARTED;\nreg  vHALTED;\nreg  vTXEV;\nreg  vLOCKUP;\nreg  vSYSRESETREQ;\nreg  vGATEHCLK;\nreg  vSLEEPING;\nreg  vSLEEPDEEP;\nreg  vWAKEUP;\nreg [33:0] vWICSENSE__33_0;\nreg  vSLEEPHOLDACKn;\nreg  vWICENACK;\nreg  vCDBGPWRUPREQ;\nreg  vSYSPWRDOWNACK;\nreg  vDBGPWRDOWNACK;\ninitial\nbegin\n    vFCLK = 0;\n    vSCLK = 0;\n    vHCLK = 0;\n    vDCLK = 0;\nend\ninteger __firstvector;\ninteger skip_output;\ninitial __firstvector = 1;\ninitial skip_output = `ARM_CM0IK_SKIPVECNUM;\ntime __capturetime;\n\ntask write_output;\nbegin\n if (__firstvector)\n   __firstvector = __firstvector-1;\n else\n begin\n   output_crf;\n   if (skip_output > 0)\n     skip_output = skip_output - 1;\n end\n __capturetime = $time;\nend\nendtask\n\n// rising FCLK\nalways @(posedge FCLK)\nbegin\n vPORESETn = PORESETn;\n vDBGRESETn = DBGRESETn;\n vHRESETn = HRESETn;\n vnTRST = nTRST;\n vHRDATA__31_0 = HRDATA[31:0];\n vHREADY = HREADY;\n vHRESP = HRESP;\n vDBGRESTART = DBGRESTART;\n vEDBGRQ = EDBGRQ;\n vNMI = NMI;\n vIRQ__31_0 = IRQ[31:0];\n vRXEV = RXEV;\n vSTCALIB__25_0 = STCALIB[25:0];\n vSTCLKEN = STCLKEN;\n vIRQLATENCY__7_0 = IRQLATENCY[7:0];\n vECOREVNUM__27_0 = ECOREVNUM[27:0];\n vSLEEPHOLDREQn = SLEEPHOLDREQn;\n vWICENREQ = WICENREQ;\n vCDBGPWRUPACK = CDBGPWRUPACK;\n vSE = SE;\n vRSTBYPASS = RSTBYPASS;\n vSYSRETAINn = SYSRETAINn;\n vSYSISOLATEn = SYSISOLATEn;\n vSYSPWRDOWN = SYSPWRDOWN;\n vDBGISOLATEn = DBGISOLATEn;\n vDBGPWRDOWN = DBGPWRDOWN;\n vHADDR__31_0 = HADDR[31:0];\n vHBURST__2_0 = HBURST[2:0];\n vHMASTLOCK = HMASTLOCK;\n vHPROT__3_0 = HPROT[3:0];\n vHSIZE__2_0 = HSIZE[2:0];\n vHTRANS__1_0 = HTRANS[1:0];\n vHWDATA__31_0 = HWDATA[31:0];\n vHWRITE = HWRITE;\n vHMASTER = HMASTER;\n vCODENSEQ = CODENSEQ;\n vCODEHINTDE__2_0 = CODEHINTDE[2:0];\n vSPECHTRANS = SPECHTRANS;\n vDBGRESTARTED = DBGRESTARTED;\n vHALTED = HALTED;\n vTXEV = TXEV;\n vLOCKUP = LOCKUP;\n vSYSRESETREQ = SYSRESETREQ;\n vGATEHCLK = GATEHCLK;\n vSLEEPING = SLEEPING;\n vSLEEPDEEP = SLEEPDEEP;\n vWAKEUP = WAKEUP;\n vWICSENSE__33_0 = WICSENSE[33:0];\n vSLEEPHOLDACKn = SLEEPHOLDACKn;\n vWICENACK = WICENACK;\n vSYSPWRDOWNACK = SYSPWRDOWNACK;\n vDBGPWRDOWNACK = DBGPWRDOWNACK;\n vSWCLKTCK = SWCLKTCK;\n vSWDITMS = SWDITMS;\n vTDI = TDI;\n vSWDO = SWDO;\n vSWDOEN = SWDOEN;\n vTDO = TDO;\n vnTDOEN = nTDOEN;\n vCDBGPWRUPREQ = CDBGPWRUPREQ;\n write_output;\n #1;\n vFCLK = FCLK;\n vSCLK = SCLK;\n vHCLK = HCLK;\n vDCLK = DCLK;\nend\n\nfunction __safedirection;\ninput directionsig;\nbegin\n  if (directionsig === 1\'bx)\n    __safedirection = 1\'b1;\n  else\n    __safedirection = directionsig;\nend\nendfunction\n\n// CRF output\nfunction [7:0] __B;\ninput direction;\ninput outvalue;\ninput invalue;\nbegin\n   if (direction === 1\'b1)\n     __B=__O(outvalue);\n   else if (direction === 1\'b0)\n     __B=__I(invalue);\n   else\n     __B=".";\nend\nendfunction\n\nfunction [7:0] __I;\ninput StrobeValue;\nbegin\n   if (StrobeValue === 1\'bz)\n     __I="u";\n   else if (StrobeValue === 1\'bx)\n     __I=`ARM_CM0IK_XVALUE;\n   else if (StrobeValue === 1\'b0)\n     __I="L";\n   else if (StrobeValue === 1\'b1)\n     __I="H";\n   else\n     __I="?";\nend\nendfunction\n\nfunction [7:0] __O;\ninput OutSignal;\nbegin\n   if (skip_output > 0)\n     __O=".";\n   else if (OutSignal === 1\'bz)\n     __O="z";\n   else if (OutSignal === 1\'bx)\n     __O=".";\n   else if (OutSignal === 1\'b0)\n     __O="0";\n   else if (OutSignal === 1\'b1)\n     __O="1";\n   else\n     __O="?";\nend\nendfunction\n\ninteger CrfHandle;\ninitial begin\n CrfHandle = $fopen("CORTEXM0INTEGRATIONIMP_vectors.crf");\n $display("%m: Writing crf to CORTEXM0INTEGRATIONIMP_vectors.crf");\n $fdisplay(CrfHandle, "#CTRM: CYCLE: FCLK (/) PRD(1000)");\n $fdisplay(CrfHandle, "#CTRM: INPUTS:");\n $fdisplay(CrfHandle, "#CTRM: FCLK <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SCLK <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: HCLK <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: DCLK <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: PORESETn <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: DBGRESETn <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: HRESETn <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SWCLKTCK <I> SWCLKTCK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: nTRST <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: HRDATA[31:0] <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: HREADY <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: HRESP <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SWDITMS <I> SWCLKTCK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: TDI <I> SWCLKTCK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: DBGRESTART <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: EDBGRQ <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: NMI <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: IRQ[31:0] <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: RXEV <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: STCALIB[25:0] <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: STCLKEN <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: IRQLATENCY[7:0] <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: ECOREVNUM[27:0] <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SLEEPHOLDREQn <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: WICENREQ <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: CDBGPWRUPACK <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SE <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: RSTBYPASS <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SYSRETAINn <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SYSISOLATEn <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: SYSPWRDOWN <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: DBGISOLATEn <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: DBGPWRDOWN <I> FCLK (/) RT0(100,600)");\n $fdisplay(CrfHandle, "#CTRM: BIDIRECTIONALS:");\n $fdisplay(CrfHandle, "#CTRM: OUTPUTS:");\n $fdisplay(CrfHandle, "#CTRM: HADDR[31:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HBURST[2:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HMASTLOCK <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HPROT[3:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HSIZE[2:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HTRANS[1:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HWDATA[31:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HWRITE <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HMASTER <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: CODENSEQ <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: CODEHINTDE[2:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SPECHTRANS <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SWDO <O> SWCLKTCK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SWDOEN <O> SWCLKTCK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: TDO <O> SWCLKTCK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: nTDOEN <O> SWCLKTCK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: DBGRESTARTED <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: HALTED <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: TXEV <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: LOCKUP <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SYSRESETREQ <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: GATEHCLK <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SLEEPING <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SLEEPDEEP <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: WAKEUP <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: WICSENSE[33:0] <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SLEEPHOLDACKn <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: WICENACK <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: CDBGPWRUPREQ <O> SWCLKTCK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: SYSPWRDOWNACK <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "#CTRM: DBGPWRDOWNACK <O> FCLK (/) CHK(400)");\n $fdisplay(CrfHandle, "-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------");\n $fdisplay(CrfHandle, "                                                                                                                                                                       S                   |                                                                                                                                                              S     S D ");\n $fdisplay(CrfHandle, "                                                                                                                                                                       L   C               |                                                                                                         D                                                    L   C Y B ");\n $fdisplay(CrfHandle, "                                                                                                                                                                       E   D       S   D   |                                                                                                         B       S                                            E   D S G ");\n $fdisplay(CrfHandle, "                                                           D                                                                            I                              E   B     S Y S B D |                                                                                             C S         G       Y                                            E   B P P ");\n $fdisplay(CrfHandle, "          D                                                B                                                                            R                            E P   G   R Y S Y G B |                                      H                                                      O P         R       S     S                                      P   G W W ");\n $fdisplay(CrfHandle, "        P B   S                                            G                                                                            Q                            C H W P   S S I S I G |                                      M                                                  C   D E         E       R G S L                                    W H W P R R ");\n $fdisplay(CrfHandle, "        O G H W                                        S   R                                                                 S S        L                            O O I W   T R S P S P |                                      A                                                H O   E C         S       E A L E                                    I O I W D D ");\n $fdisplay(CrfHandle, "        R R R C                                  H H   W   E E                                                               T T        A                            R L C R   B E O W O W |                                    H S           H                                H H M D   H H   S   n T H   L S T E E W                                  C L C R O O ");\n $fdisplay(CrfHandle, "        E E E L n                                R R H D   S D                                                               C C        T                            E D E U   Y T L R L R |                                H   B T    H   H  T                                W W A E   I T   W   T A A   O E E E P A                                  S D E U W W ");\n $fdisplay(CrfHandle, "F S H D S S S K T                                D E R I   T B                                    R                          A L        E                            V R N P   P A A D A D |                                A   U L    P   S  R                                D R S N   N R S D   D R L T C T H P D K                                  E A N P N N ");\n $fdisplay(CrfHandle, "C C C C E E E T R                                A A E T T A G N                                I X                          L K        N                            N E R A   A I T O T O |                                D   R O    R   I  A                                A I T S   T A W O T O T T X K R C I E E                                  N C A R A A ");\n $fdisplay(CrfHandle, "L L L L T T T C S                                T D S M D R R M                                R E                          I E        C                            U Q E C S S N E W E W |                                D   S C    O   Z  N                                T T E E   D N D E D E E E E U E L N E U                                  S K C E C C ");\n $fdisplay(CrfHandle, "K K K K n n n K T                                A Y P S I T Q I                                Q V                          B N        Y                            M n Q K E S n n N n N |                                R   T K    T   E  S                                A E R Q   E S O N O N D D V P Q K G P P                                  E n K Q K K ");\n $fdisplay(CrfHandle, "-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------");\n $fdisplay(CrfHandle, "                  3322222222221111111111                         3322222222221111111111             2222221111111111                      222222221111111111                               | 3322222222221111111111                             3322222222221111111111                                                 333322222222221111111111                     ");\n $fdisplay(CrfHandle, "                  10987654321098765432109876543210               10987654321098765432109876543210   54321098765432109876543210   76543210 7654321098765432109876543210                     | 10987654321098765432109876543210 210   3210 210 10 10987654321098765432109876543210       210                             3210987654321098765432109876543210           ");\n $fdisplay(CrfHandle, "-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------");\nend\n\ntask output_crf;\nbegin\n $fdisplay(CrfHandle, "%s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s | %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s %s } %t", \n  __I(vFCLK),\n  __I(vSCLK),\n  __I(vHCLK),\n  __I(vDCLK),\n  __I(vPORESETn),\n  __I(vDBGRESETn),\n  __I(vHRESETn),\n  __I(vSWCLKTCK),\n  __I(vnTRST),{\n  __I(vHRDATA__31_0[31]),\n  __I(vHRDATA__31_0[30]),\n  __I(vHRDATA__31_0[29]),\n  __I(vHRDATA__31_0[28]),\n  __I(vHRDATA__31_0[27]),\n  __I(vHRDATA__31_0[26]),\n  __I(vHRDATA__31_0[25]),\n  __I(vHRDATA__31_0[24]),\n  __I(vHRDATA__31_0[23]),\n  __I(vHRDATA__31_0[22]),\n  __I(vHRDATA__31_0[21]),\n  __I(vHRDATA__31_0[20]),\n  __I(vHRDATA__31_0[19]),\n  __I(vHRDATA__31_0[18]),\n  __I(vHRDATA__31_0[17]),\n  __I(vHRDATA__31_0[16]),\n  __I(vHRDATA__31_0[15]),\n  __I(vHRDATA__31_0[14]),\n  __I(vHRDATA__31_0[13]),\n  __I(vHRDATA__31_0[12]),\n  __I(vHRDATA__31_0[11]),\n  __I(vHRDATA__31_0[10]),\n  __I(vHRDATA__31_0[9]),\n  __I(vHRDATA__31_0[8]),\n  __I(vHRDATA__31_0[7]),\n  __I(vHRDATA__31_0[6]),\n  __I(vHRDATA__31_0[5]),\n  __I(vHRDATA__31_0[4]),\n  __I(vHRDATA__31_0[3]),\n  __I(vHRDATA__31_0[2]),\n  __I(vHRDATA__31_0[1]),\n  __I(vHRDATA__31_0[0])},\n  __I(vHREADY),\n  __I(vHRESP),\n  __I(vSWDITMS),\n  __I(vTDI),\n  __I(vDBGRESTART),\n  __I(vEDBGRQ),\n  __I(vNMI),{\n  __I(vIRQ__31_0[31]),\n  __I(vIRQ__31_0[30]),\n  __I(vIRQ__31_0[29]),\n  __I(vIRQ__31_0[28]),\n  __I(vIRQ__31_0[27]),\n  __I(vIRQ__31_0[26]),\n  __I(vIRQ__31_0[25]),\n  __I(vIRQ__31_0[24]),\n  __I(vIRQ__31_0[23]),\n  __I(vIRQ__31_0[22]),\n  __I(vIRQ__31_0[21]),\n  __I(vIRQ__31_0[20]),\n  __I(vIRQ__31_0[19]),\n  __I(vIRQ__31_0[18]),\n  __I(vIRQ__31_0[17]),\n  __I(vIRQ__31_0[16]),\n  __I(vIRQ__31_0[15]),\n  __I(vIRQ__31_0[14]),\n  __I(vIRQ__31_0[13]),\n  __I(vIRQ__31_0[12]),\n  __I(vIRQ__31_0[11]),\n  __I(vIRQ__31_0[10]),\n  __I(vIRQ__31_0[9]),\n  __I(vIRQ__31_0[8]),\n  __I(vIRQ__31_0[7]),\n  __I(vIRQ__31_0[6]),\n  __I(vIRQ__31_0[5]),\n  __I(vIRQ__31_0[4]),\n  __I(vIRQ__31_0[3]),\n  __I(vIRQ__31_0[2]),\n  __I(vIRQ__31_0[1]),\n  __I(vIRQ__31_0[0])},\n  __I(vRXEV),{\n  __I(vSTCALIB__25_0[25]),\n  __I(vSTCALIB__25_0[24]),\n  __I(vSTCALIB__25_0[23]),\n  __I(vSTCALIB__25_0[22]),\n  __I(vSTCALIB__25_0[21]),\n  __I(vSTCALIB__25_0[20]),\n  __I(vSTCALIB__25_0[19]),\n  __I(vSTCALIB__25_0[18]),\n  __I(vSTCALIB__25_0[17]),\n  __I(vSTCALIB__25_0[16]),\n  __I(vSTCALIB__25_0[15]),\n  __I(vSTCALIB__25_0[14]),\n  __I(vSTCALIB__25_0[13]),\n  __I(vSTCALIB__25_0[12]),\n  __I(vSTCALIB__25_0[11]),\n  __I(vSTCALIB__25_0[10]),\n  __I(vSTCALIB__25_0[9]),\n  __I(vSTCALIB__25_0[8]),\n  __I(vSTCALIB__25_0[7]),\n  __I(vSTCALIB__25_0[6]),\n  __I(vSTCALIB__25_0[5]),\n  __I(vSTCALIB__25_0[4]),\n  __I(vSTCALIB__25_0[3]),\n  __I(vSTCALIB__25_0[2]),\n  __I(vSTCALIB__25_0[1]),\n  __I(vSTCALIB__25_0[0])},\n  __I(vSTCLKEN),{\n  __I(vIRQLATENCY__7_0[7]),\n  __I(vIRQLATENCY__7_0[6]),\n  __I(vIRQLATENCY__7_0[5]),\n  __I(vIRQLATENCY__7_0[4]),\n  __I(vIRQLATENCY__7_0[3]),\n  __I(vIRQLATENCY__7_0[2]),\n  __I(vIRQLATENCY__7_0[1]),\n  __I(vIRQLATENCY__7_0[0])},{\n  __I(vECOREVNUM__27_0[27]),\n  __I(vECOREVNUM__27_0[26]),\n  __I(vECOREVNUM__27_0[25]),\n  __I(vECOREVNUM__27_0[24]),\n  __I(vECOREVNUM__27_0[23]),\n  __I(vECOREVNUM__27_0[22]),\n  __I(vECOREVNUM__27_0[21]),\n  __I(vECOREVNUM__27_0[20]),\n  __I(vECOREVNUM__27_0[19]),\n  __I(vECOREVNUM__27_0[18]),\n  __I(vECOREVNUM__27_0[17]),\n  __I(vECOREVNUM__27_0[16]),\n  __I(vECOREVNUM__27_0[15]),\n  __I(vECOREVNUM__27_0[14]),\n  __I(vECOREVNUM__27_0[13]),\n  __I(vECOREVNUM__27_0[12]),\n  __I(vECOREVNUM__27_0[11]),\n  __I(vECOREVNUM__27_0[10]),\n  __I(vECOREVNUM__27_0[9]),\n  __I(vECOREVNUM__27_0[8]),\n  __I(vECOREVNUM__27_0[7]),\n  __I(vECOREVNUM__27_0[6]),\n  __I(vECOREVNUM__27_0[5]),\n  __I(vECOREVNUM__27_0[4]),\n  __I(vECOREVNUM__27_0[3]),\n  __I(vECOREVNUM__27_0[2]),\n  __I(vECOREVNUM__27_0[1]),\n  __I(vECOREVNUM__27_0[0])},\n  __I(vSLEEPHOLDREQn),\n  __I(vWICENREQ),\n  __I(vCDBGPWRUPACK),\n  __I(vSE),\n  __I(vRSTBYPASS),\n  __I(vSYSRETAINn),\n  __I(vSYSISOLATEn),\n  __I(vSYSPWRDOWN),\n  __I(vDBGISOLATEn),\n  __I(vDBGPWRDOWN),{\n  __O(vHADDR__31_0[31]),\n  __O(vHADDR__31_0[30]),\n  __O(vHADDR__31_0[29]),\n  __O(vHADDR__31_0[28]),\n  __O(vHADDR__31_0[27]),\n  __O(vHADDR__31_0[26]),\n  __O(vHADDR__31_0[25]),\n  __O(vHADDR__31_0[24]),\n  __O(vHADDR__31_0[23]),\n  __O(vHADDR__31_0[22]),\n  __O(vHADDR__31_0[21]),\n  __O(vHADDR__31_0[20]),\n  __O(vHADDR__31_0[19]),\n  __O(vHADDR__31_0[18]),\n  __O(vHADDR__31_0[17]),\n  __O(vHADDR__31_0[16]),\n  __O(vHADDR__31_0[15]),\n  __O(vHADDR__31_0[14]),\n  __O(vHADDR__31_0[13]),\n  __O(vHADDR__31_0[12]),\n  __O(vHADDR__31_0[11]),\n  __O(vHADDR__31_0[10]),\n  __O(vHADDR__31_0[9]),\n  __O(vHADDR__31_0[8]),\n  __O(vHADDR__31_0[7]),\n  __O(vHADDR__31_0[6]),\n  __O(vHADDR__31_0[5]),\n  __O(vHADDR__31_0[4]),\n  __O(vHADDR__31_0[3]),\n  __O(vHADDR__31_0[2]),\n  __O(vHADDR__31_0[1]),\n  __O(vHADDR__31_0[0])},{\n  __O(vHBURST__2_0[2]),\n  __O(vHBURST__2_0[1]),\n  __O(vHBURST__2_0[0])},\n  __O(vHMASTLOCK),{\n  __O(vHPROT__3_0[3]),\n  __O(vHPROT__3_0[2]),\n  __O(vHPROT__3_0[1]),\n  __O(vHPROT__3_0[0])},{\n  __O(vHSIZE__2_0[2]),\n  __O(vHSIZE__2_0[1]),\n  __O(vHSIZE__2_0[0])},{\n  __O(vHTRANS__1_0[1]),\n  __O(vHTRANS__1_0[0])},{\n  __O(vHWDATA__31_0[31]),\n  __O(vHWDATA__31_0[30]),\n  __O(vHWDATA__31_0[29]),\n  __O(vHWDATA__31_0[28]),\n  __O(vHWDATA__31_0[27]),\n  __O(vHWDATA__31_0[26]),\n  __O(vHWDATA__31_0[25]),\n  __O(vHWDATA__31_0[24]),\n  __O(vHWDATA__31_0[23]),\n  __O(vHWDATA__31_0[22]),\n  __O(vHWDATA__31_0[21]),\n  __O(vHWDATA__31_0[20]),\n  __O(vHWDATA__31_0[19]),\n  __O(vHWDATA__31_0[18]),\n  __O(vHWDATA__31_0[17]),\n  __O(vHWDATA__31_0[16]),\n  __O(vHWDATA__31_0[15]),\n  __O(vHWDATA__31_0[14]),\n  __O(vHWDATA__31_0[13]),\n  __O(vHWDATA__31_0[12]),\n  __O(vHWDATA__31_0[11]),\n  __O(vHWDATA__31_0[10]),\n  __O(vHWDATA__31_0[9]),\n  __O(vHWDATA__31_0[8]),\n  __O(vHWDATA__31_0[7]),\n  __O(vHWDATA__31_0[6]),\n  __O(vHWDATA__31_0[5]),\n  __O(vHWDATA__31_0[4]),\n  __O(vHWDATA__31_0[3]),\n  __O(vHWDATA__31_0[2]),\n  __O(vHWDATA__31_0[1]),\n  __O(vHWDATA__31_0[0])},\n  __O(vHWRITE),\n  __O(vHMASTER),\n  __O(vCODENSEQ),{\n  __O(vCODEHINTDE__2_0[2]),\n  __O(vCODEHINTDE__2_0[1]),\n  __O(vCODEHINTDE__2_0[0])},\n  __O(vSPECHTRANS),\n  __O(vSWDO),\n  __O(vSWDOEN),\n  __O(vTDO),\n  __O(vnTDOEN),\n  __O(vDBGRESTARTED),\n  __O(vHALTED),\n  __O(vTXEV),\n  __O(vLOCKUP),\n  __O(vSYSRESETREQ),\n  __O(vGATEHCLK),\n  __O(vSLEEPING),\n  __O(vSLEEPDEEP),\n  __O(vWAKEUP),{\n  __O(vWICSENSE__33_0[33]),\n  __O(vWICSENSE__33_0[32]),\n  __O(vWICSENSE__33_0[31]),\n  __O(vWICSENSE__33_0[30]),\n  __O(vWICSENSE__33_0[29]),\n  __O(vWICSENSE__33_0[28]),\n  __O(vWICSENSE__33_0[27]),\n  __O(vWICSENSE__33_0[26]),\n  __O(vWICSENSE__33_0[25]),\n  __O(vWICSENSE__33_0[24]),\n  __O(vWICSENSE__33_0[23]),\n  __O(vWICSENSE__33_0[22]),\n  __O(vWICSENSE__33_0[21]),\n  __O(vWICSENSE__33_0[20]),\n  __O(vWICSENSE__33_0[19]),\n  __O(vWICSENSE__33_0[18]),\n  __O(vWICSENSE__33_0[17]),\n  __O(vWICSENSE__33_0[16]),\n  __O(vWICSENSE__33_0[15]),\n  __O(vWICSENSE__33_0[14]),\n  __O(vWICSENSE__33_0[13]),\n  __O(vWICSENSE__33_0[12]),\n  __O(vWICSENSE__33_0[11]),\n  __O(vWICSENSE__33_0[10]),\n  __O(vWICSENSE__33_0[9]),\n  __O(vWICSENSE__33_0[8]),\n  __O(vWICSENSE__33_0[7]),\n  __O(vWICSENSE__33_0[6]),\n  __O(vWICSENSE__33_0[5]),\n  __O(vWICSENSE__33_0[4]),\n  __O(vWICSENSE__33_0[3]),\n  __O(vWICSENSE__33_0[2]),\n  __O(vWICSENSE__33_0[1]),\n  __O(vWICSENSE__33_0[0])},\n  __O(vSLEEPHOLDACKn),\n  __O(vWICENACK),\n  __O(vCDBGPWRUPREQ),\n  __O(vSYSPWRDOWNACK),\n  __O(vDBGPWRDOWNACK), __capturetime);\nend\nendtask\nendmodule\n'
[module_definition, module_output_code, module_definition_with_upper_comments] = module_extraction(code)
print(code)
```
